```python
#Importing the libraries
import pandas as pd 
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

data = pd.read_csv("D://Data/propulsion.csv")
data.head()
```

```python
data.info()
```

```python
data.isnull().sum()
```

```python
data.drop(["Unnamed: 0"],axis = 1, inplace = True)
```

```python
data.columns
```

```python
NetF = list(data.columns)
```

```python
NetF
```

```python
dat = pd.read_csv("D://Data/propulsion.csv")
#Two temperary lists/arrays for our use
temp1=[]
temp2=[]

#Fiding Features whose skewness can be decreases
skewFeature=[]

for i in NetF:
    temp1.append(dat[i].skew())
    if (dat[i].min()) > 0 :
        dat[i] = dat[i].map(lambda s: np.log(s)if s > 0 else 0)
    #To check     
    if (dat[i].min()) <= 0 :
        dat[i] = dat[i].map(lambda s: -np.power(-s, 1./3) if s < 0 else np.power(s, 1./3)) 
        
    temp2.append(dat[i].skew())
for n in range(4):
    if abs(temp1[n])>abs(temp2[n]):
        skewFeature.append(NetF[n])
print(skewFeature)
```

```python
for i in skewFeature:
    if (data[i].min()) > 0 :
        data[i] = data[i].map(lambda k: np.log(k)if k > 0 else 0)
    #Checking
    if (data[i].min()) <= 0 :
        data[i] = data[i].map(lambda x: -np.power(-x, 1./3) if x < 0 else np.power(x, 1./3)) 
        
    print(i,"\t\t",data[i].skew())
```

```python
#thing about outliers is that its best to remove them or else they mess things up

Ps = data.quantile(0.25)
Ps1 = data.quantile(0.75)
result = Ps1 - Ps

#Showing results
print(result)

data = data[~((data < (Ps - 2 * result)) |(data > (Ps1 + 2 * result))).any(axis=1)]
data.shape
```

```python
plt.figure(figsize=(26, 12))
sns.heatmap(data.corr(),cmap="OrRd",annot=True)
plt.show()
```

```python
from sklearn.model_selection import train_test_split
data_train, data_test = train_test_split(data, train_size = 0.8, test_size = 0.2, random_state = 0 )
```

```python
from sklearn.preprocessing import StandardScaler,scale
#we create an object of the class StandardScaler
sc = StandardScaler() 

col_to_scale = NetF

data_train[col_to_scale] = sc.fit_transform(data_train[col_to_scale])
data_test[col_to_scale] = sc.fit_transform(data_test[col_to_scale])

data_train.head()
```

```python
y_train = data_train.loc[:,data_train.columns == 'GT Turbine decay state coefficient.']
X_train = data_train.loc[:, data_train.columns != 'GT Turbine decay state coefficient.']
```

```python
y_test = data_test.loc[:,data_test.columns == 'GT Turbine decay state coefficient.']
X_test = data_test.loc[:, data_test.columns != 'GT Turbine decay state coefficient.']
```

```python
X_train.columns
```

```python
from keras.models import Sequential
from keras.layers import Dense
#create model

model = Sequential()
#input layer
model.add(Dense(input_dim = 17, units = 256, kernel_initializer='normal', activation='relu'))

#6 Dense Layers
model.add(Dense(128,activation = 'relu', kernel_initializer='normal'))

#Output layer with softmax
model.add(Dense(1, kernel_initializer='normal' ))
```

```python
model.compile(optimizer='adam', loss='mean_squared_error')
```

```python
model.fit(X_train, y_train, validation_data=(X_test, y_test), epochs=40, batch_size = 10)
```

```python
y_pred = model.predict(X_test)
```

```python
resid = y_test - y_pred
```

```python
from statsmodels.stats.stattools import durbin_watson
print(durbin_watson(resid))
```

```python
from sklearn.metrics import r2_score 
r2_score(y_test, y_pred)
```

```python
from sklearn.decomposition import PCA
pca = PCA(0.96)
pca.fit(X_train)
print(pca.n_components_)
```

```python
X_train = pca.transform(X_train)
X_test = pca.transform(X_test)
```

```python
# Making predictions
import statsmodels.api as sm 

X_train = sm.add_constant(X_train)
print(X_train.shape)
lm = sm.OLS(y_train,X_train).fit()

X_test = sm.add_constant(X_test)
y_pred=lm.predict(X_test)
```
